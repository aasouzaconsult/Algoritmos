### Carregando a base de dados

BigMart Sales Prediction practice problem

- `Item_Identifier`: ID exclusivo do produto
- `Item_Weight`: Peso do produto
- `Item_Fat_Content`: Se o produto é baixo teor de gordura ou não
- `Item_Visibility`: A % da área total de exibição de todos os produtos em uma loja alocada para o produto específico
- `Item_Type`: A categoria à qual o produto pertence
- `Item_MRP`: Preço máximo de varejo (preço de tabela) do produto
- `Outlet_Identifier`: ID exclusivo da loja
- `Outlet_Establishment_Year`: O ano em que a loja foi criada
- `Outlet_Size`: O tamanho da loja em termos de área coberta
- `Outlet_Location_Type`: O tipo de cidade em que a loja está localizada
- `Outlet_Type`: Se a saída é apenas uma mercearia ou algum tipo de supermercado
- `Item_Outlet_Sales`: Vendas do produto na loja particular. Esta é a variável de resultado a ser prevista.

```
df = pd.read_csv('/content/drive/MyDrive/3. Estudos/Profissional/IA, ML/0. Scripts/Dados/train.csv')
```

```
df.head()
```

### Selecionando os atributos

Iremos selecionar 3 atributos
- `Outlet_Establishment_Year`
  - ano que a loja foi criada
- `Item_MRP`
  - preço máximo do produto
- `Item_Outlet_Sales` (essa é o que queremos prever)

```
df[['Outlet_Establishment_Year','Item_MRP','Item_Outlet_Sales']][:5]
```

```
# Variaveis Indenpendentes ou Preditoras
X = df[['Outlet_Establishment_Year','Item_MRP']]
```

#### Separando os conjuntos de treino e teste


```
# y = df.Item_Outlet_Sales - variável dependente - alvo
x_train, x_test, y_train, y_test = train_test_split( X, df.Item_Outlet_Sales, test_size=0.3)
```

#### Verificando o shape dos dados

```
x_train.shape, y_train.shape
```

```
x_test.shape, y_test.shape
```

### Instânciando a Regressão Linear

```
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
```

### Treinando o modelo

```
lreg.fit(x_train,y_train)
```

###  Predizendo os valores para o conjunto de teste

```
pred = lreg.predict(x_test)
```

###  Calculando o MSE (Mean Square Error)
A ideia da regressão é diminuir o erro...

```
mse = np.mean((pred - y_test)**2)
mse
```

### Calculando os Coeficientes


```
coeff = DataFrame(x_train.columns)
coeff['Coeficientes'] = Series(lreg.coef_)
coeff
```

### Calculando o r-squared

```
lreg.score(x_test,y_test)
```

### Regressão Linear com mais variáveis.

#### Verificando o atributo Item Weight

```
df['Item_Weight'].head()
```

#### Verificando e tratando registros nulos

```
# 1463 registros nulos para a variável Item_Weight - Peso Produto
df['Item_Weight'].isnull().sum()
```

```
# Onde for nulo vai preencher com a média neste caso
df['Item_Weight'].fillna((df['Item_Weight'].mean()), inplace=True)
```

#### Selecionando novamente os 3 atributos
Agora tb com o Item_Weight

```
X = df.loc[:,['Outlet_Establishment_Year','Item_MRP','Item_Weight']]
```

#### Separando os conjuntos de treino e teste

```
x_train, x_test, y_train, y_test = train_test_split( X, df.Item_Outlet_Sales, test_size=0.3)
```

#### Treinando novamente o modelo

```
lreg.fit(x_train,y_train)
```

#### Executando a predição e calculando o MSE

```
pred = lreg.predict(x_test)
mse = np.mean((pred - y_test)**2)
mse
```

Observem que o erro diminuiu, em relação ao primeiro exemplo.

#### Verificando os coeficientes

```
coeff = DataFrame(x_train.columns)
coeff['Coeficientes'] = Series(lreg.coef_)
coeff
```

#### Calculando o r-squared

```
lreg.score(x_test,y_test)
```
