### Relação de Variáveis

**Função** relplot()

Método de alto nível para visualização de relaçoes estatísticas entre variáveis. O método relplot() já combina o **scatter plot** e **line plots**. 

Muito útil para ver relações entre duas ou mais variáveis com o uso de parâmetros de semântica dos dados.

**Visualizando a relação entre a variável valor total da conta e gorjeta.**

```python
tips = sns.load_dataset("tips")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips);
```

**Visualizando a relação entre as variávels total da conta e gorjeta e colorindo cada ponto com a variável smoker(fumante)**

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

**Visualizando a relação entre as variávels total da conta e gorjeta e colorindo cada ponto com a variável sex**



```python
sns.relplot(x="total_bill", y="tip", hue="sex", data=tips);
```

**Visualizando a relação entre as variávels total da conta e gorjeta e colorindo cada ponto com a variável  sexo e definindo um estilo também baseado na variável sexo**

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="sex", data=tips);
```

**É possível representar quatro variáveis alterando a tonalidade e estilo de cada ponto. Mas é preciso cautela, pois, a visualização pode ficar confusa.**

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", data=tips);
```

**Em casos de variáveis numéricas para definir a tonalidade a paleta de cores altera de forma seguencial.**

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
```

**É possível customizar a paleta de cores sequenciais no link**: 

* https://seaborn.pydata.org/generated/seaborn.cubehelix_palette.html#seaborn.cubehelix_palette

**Usando a opção size para definir o tamanho de cada ponto baseado em alguma coluna da base de dados**

```python
sns.relplot(x="total_bill", y="tip", size="size", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15,300), data=tips);
```

### Cores no Seaborn

* A funcao que controla as cores no seaborn e a color_palette(). 
* Essa funcao fornece uma interface para gerar cores no seaborn.
* Usamos esse metodo internamente ao usar o parametro palette.
* O metodo color_palette() recebe o nome de uma paleta de cores do seaborn ou um colormap do matplotlib
* O metodo set_palette() recebe os mesmos parametros e eh usado para definir um estilo padrao para todos os plots.
* As cores devem ser definidas levando em consideracao o tipo de dado que queremos plotar. (qualitativos, sequenciais,diverging)


**Ferramenta Color Brewer 2.0 **

*  http://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3

**Paleta de cores para dados Quantitativos**

```python
sns.palplot(sns.color_palette("Accent"))
```

**Paleta de cores para dados divergentes**

```python
sns.palplot(sns.color_palette("PuOr"))
```

**Paleta de cores para dados sequenciais**

```python
sns.palplot(sns.color_palette("Blues"))
```

**Gráfico usando a paleta de cor sequencial "Blues"**

```python
sns.relplot(x="total_bill", y="tip", hue="size", palette="Blues", size="size", sizes=(12,250), data=tips);
```

**Gráfico usando a paleta de cor qualitativa "Set2"**

```python
sns.catplot(x="sex", kind="count", palette="Set2", data=tips);
```

**Gráfico usando a paleta de cor divergente "PiYG"**

```python
sns.heatmap(tips.corr(), annot=True, cmap="PiYG")
```

**Definindo cores padrão para gráficos Searborn**

```python
sns.set_palette("pastel")

# plot usando a paleta de cor "pastel"
sns.relplot(x="total_bill", y="tip", data=tips);
```

**Instrução with para definir cor temporariamente**

```python
# Plota o gráfico definindo uma opção de cor temporária
with sns.color_palette("Accent"):
  sns.relplot(x="total_bill", y="tip", size="size", sizes=(12,250), data=tips);
```

**Gráficos de barras ou também conhecidos como gráficos de colunas**

```python
sns.catplot(x="sex", kind="count", palette="Set2", data=tips);
```

** Gráfico de barras horizontais usando a coluna *day* **

```python
sns.catplot(y="day", kind="count", palette="Set1", data=tips);
```

### Plotando dados categóricos

* Quando trabalhamos com variáveis categóricas temos que visualizar dados de formas diferentes.
* O Seaborn fornece o método catplot() que já contém diversos tipos de gráficos embutidos.
* Isso facilita, pois, você pode usar diferentes gráficos usando um mesmo método.


**Gráficos de barras ou também conhecidos como gráficos de colunas**

```python
sns.catplot(x="sex", kind="count", palette="Set2", data=tips);
```

** Gráfico de barras horizontais usando a coluna *day* **

```python
sns.catplot(y="day", kind="count", palette="Set1", data=tips);
```

**Scatter plot com dados categóricos**

```python
sns.catplot(x="day", y="total_bill", palette="Set2", data=tips);
```

**O parâmetro swarm evita sobreposição de pontos**

```python
sns.catplot(x="day", y="total_bill", kind='swarm', data=tips);
```

**O parâmetro hue permite adicionarmos uma terceira variável a nossa visualização**

```python
sns.catplot(x="day", y="total_bill", kind='swarm', hue="sex", palette="Dark2", data=tips);
```

* O parametro **order** permite alterarmos a ordem padrão das categorias que estão sendo exibidas.

* Isso é útil quando temos mais um gráfico na mesma figura e queremos manter as mesmas ordens.

```python
sns.catplot(x="day", y="total_bill", kind='swarm', hue="sex", order=['Sat','Sun','Fri','Thur'], data=tips);
```

**Invertendo a visualização para plots horizontais**

```python
sns.catplot(x="total_bill", y="day", kind="swarm", data=tips);
```

### Gráficos com Regressão

* Quando temos muitas variáveis quantitativas em nossos dados é interessante visualizar como estas se relacionam.
* Podemos visualizar essas informações com linhas de regressão.
* Com modelos de regressão simples é possível checar se existe alguma **correlacão** entre algumas variáveis.

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

**Exibe a linha de regressão para visualizar a dispersão entre as variáveis**

```python
%matplotlib inline
sns.lmplot(x="total_bill", y="tip", hue="sex", palette="Pastel1", data=tips);
```

** Usando o parâmetro col para segregar os gráficos pelo valor da coluna categórica *time* **

```python
%matplotlib inline
sns.lmplot(x="total_bill",y="tip", hue="smoker", col="time", palette="Set2", data=tips);
```

**Parâmetros scatter_kws e line_kws**

```python
%matplotlib inline
sns.lmplot(x="total_bill",
           y="tip", 
           hue="smoker", 
           col="time", 
           palette="Set1",
           data=tips, 
           scatter_kws={'s': 90, 'alpha': 0.5},
           line_kws={'lw': 4});
```

** Segregando gráficos pelo parâmetro *col* e pelo parâmetro *row* **

```python
%matplotlib inline
sns.lmplot(x="total_bill", y="tip", hue="smoker",col="time", row="sex", palette="Set2", data=tips);
```

#### Visualizando Outliers

**Boxplot de Dias da semana por total de conta**

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips);
```

** Gráfico de Boxplot dos dias por total de conta e se a pessoa é fumante ou não.**

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips);
```

* Tipo de Boxplot com foco maior na forma dos dados do que nos outliers.
* Esse tipo de gráfico é interessante quando temos grandes datasets.

```python
sns.catplot(x="day", y="total_bill", kind="boxen", data=tips);
```

* O Método stripplot permite plotar a distribuição dos dados.
* Podemos combinar os dois gráficos para ter mais informação.

```python
sns.catplot(x="day", y="total_bill", kind="boxen", data=tips);
sns.stripplot(x="day", y="total_bill", data=tips, color="blue");
```

**O método catplot() permite usarmos facetgrid assim podemos combinar gráficos em uma única figura**

```python
%matplotlib inline
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=tips, kind="boxen",height=4);
```
